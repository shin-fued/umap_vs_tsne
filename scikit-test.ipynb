```python
from sklearn.manifold import TSNE, LocallyLinearEmbedding, Isomap, MDS, SpectralEmbedding
import umap
from sklearn.decomposition import PCA
```

```python
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
dia = pd.read_csv("diabetes.csv")
reducer = umap.UMAP()
len(dia.columns)
```

```python
scaled_dia_data = StandardScaler().fit_transform(dia.values[:-1])
```

```python
embeddings = [reducer.fit_transform(scaled_dia_data) for _ in range(10)]
```

```python
for embedding in embeddings:
    plt.scatter(
    embedding[:, 0],
    embedding[:, 1])
    plt.show()
```

```python
X_embeddeds = [TSNE(n_components=2, learning_rate='auto',
                  init='random', perplexity=100).fit_transform(dia.values) for _ in range(10)]
```

```python
for X_embedded in X_embeddeds:
    plt.scatter(
    X_embedded[:, 0],
    X_embedded[:, 1])
    plt.show()
```
